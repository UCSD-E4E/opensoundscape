```python
import pandas as pd
import numpy as np
```

```python
ibp = pd.read_csv('./ibp-alpha-codes.csv',header=None)
ibp.columns = columns=['common','alpha','scientific','scicode']
ibp
```

```python
species_table = pd.read_csv('./species_table.csv')
species_table
```

```python
species_table = species_table.set_index('scientific')
```

```python
for i,row in ibp.iterrows():
    sci_lower = row.scientific.replace(' ','-').lower()
    if sci_lower in species_table.index:
        species_table.at[sci_lower,'ibp_alpha']=row.alpha.replace('*','')
species_table
```
